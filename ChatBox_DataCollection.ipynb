```python
from pprint import pprint
import random
import nltk

# Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes]
# Body Mass Index (BMI)
# Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week
# categorical
# Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good? (0-30 days)
# Thinking about your mental health, for how many days during the past 30 days was your mental health not good? (0-30 days)
# Are you male or female?
# Imputed race/ethnicity value
# Binary
# (Ever told) (you had) diabetes?
# (Ever told) (you had) asthma?
# (Ever told) (you had) skin cancer?
# (Ever told) (you had) a stroke?
# Not including kidney stones, bladder infection or incontinence, were you ever told you had kidney disease?
# Do you have serious difficulty walking or climbing stairs?
# Adults who reported doing physical activity or exercise during the past 30 days other than their regular job
# Would you say that in general your health is...
# numerical
# On average, how many hours of sleep do you get in a 24-hour period?
# Fourteen-level age category
```

```python
#design question and categorize questions based on the type of answer;
answerList = []
```

```python
def StringQuestion():
    stringQuestion = {("difficulty walking or climbing stairs","?"),("do any exercise in past 30 days","?")}
    unknownsStQi = { (fact,value) for (fact,value) \
                 in stringQuestion if value=="?" }
    active = True
    while active:
        if unknownsStQi: #is non-empty
            fact, value = random.choice(list(unknownsStQi))
            question = "Do you have"+fact+"? "+" answer must be yes or no"
#             print(question)
            reply = input(question)
            if reply=="bye":
                active = False
                continue
            tokens = nltk.word_tokenize(reply)
    #         tagged = nltk.pos_tag(tokens)
            targetwords = [ word for word in tokens if (word=="yes"or word == "no")]
            if(targetwords):
                unknownsStQi.remove((fact,value))
#                 print(targetwords)
                answerList.append((fact,targetwords[0]))
#                 print(answerList)
                if (not unknownsStQi) :
                    active = False
    print("code end")
# StringQuestion()
```

```python
def CategoricalQuestion():
    categoricalQuestion = {("mental health","?"),("physical health","?")}
    unknownsCateQi = { (fact,value) for (fact,value) \
                 in categoricalQuestion if value=="?" }
    active = True
    while active:
        if unknownsCateQi: #is non-empty
            fact, value = random.choice(list(unknownsCateQi))
            question = "How many days you think your "+fact+"is good in last 30 days? "+\
            "answer must be number <= 30 "
#             print(question)
            reply = input(question)
            if reply=="bye":
                active = False
                continue
            tokens = nltk.word_tokenize(reply)
    #         tagged = nltk.pos_tag(tokens)
            targetwords = [ word for word in tokens if (float(word)>=0 and float(word)<= 30)]
            if(targetwords):
                unknownsCateQi.remove((fact,value))
#                 print(targetwords)
                answerList.append((fact,targetwords[0]))
#                 print(answerList)
                if (not unknownsCateQi) :
                    active = False
    print("code end")
```

```python
def BinaryQuestion():
    binaryQuestion = { ("diabetes", "?"), ("ashma", "?"),\
                  ("skin cancer", "?"),("stroke", "?"),\
                  ("kidney disease", "?")}
    unknownsBiQues = { (fact,value) for (fact,value) \
                 in binaryQuestion if value=="?" }
    active = True
    while active:
        if unknownsBiQues: #is non-empty
            fact, value = random.choice(list(unknownsBiQues))
            question = "Have you ever had "+fact+"? "+" answer must be yes or no"
#             print(question)
            reply = input(question)
            if reply=="bye":
                active = False
                continue
            tokens = nltk.word_tokenize(reply)
    #         tagged = nltk.pos_tag(tokens)
            targetwords = [ word for word in tokens if (word=="yes"or word == "no")]
            if(targetwords):
                unknownsBiQues.remove((fact,value))
#                 print(targetwords)
                answerList.append((fact,targetwords[0]))
#                 print(answerList)
                if (not unknownsBiQues) :
                    active = False
    print("code end")
```

```python
def NumerQuestion():
    numericalQuestion = {("sleep" , "?"), ("age","?"),("drink","?"),("cigarrate","?")}
    unknownsNumQi = { (fact,value) for (fact,value) \
                 in numericalQuestion if value=="?" }
    active = True
    while active:
        if unknownsNumQi: #is non-empty
            fact, value = random.choice(list(unknownsNumQi))
            if(fact == "sleep"):
                question = "How many hours do you sleep in 24hours period "+\
                "answer must be number"
            if(fact == "age"):
                question = "How old are you?"+\
                "answer must be number"
            if(fact == "drink"):
                question = "How many bottles of wine do you drink per week? "+\
                "answer must be number"
            if(fact == "cigarrate"):
                question = "How many cigarrate do you smoke per week? "+\
                "answer must be number"
#             print(question)
            reply = input(question)
            if reply=="bye":
                active = False
                continue
            tokens = nltk.word_tokenize(reply)
    #         tagged = nltk.pos_tag(tokens)
            targetwords = [ word for word in tokens if float(word)]
            if(targetwords):
                unknownsNumQi.remove((fact,value))
                print(targetwords)
                answerList.append((fact,targetwords[0]))
#                 print(answerList)
                if (not unknownsNumQi) :
                    active = False
    print("code end")
```
